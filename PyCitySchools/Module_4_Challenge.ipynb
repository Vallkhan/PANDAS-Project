```python
import pandas as pd
import csv 

schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")

schools_df.head()

```

```python
#Total Unique Schools
total_unique_schools = schools_df["School ID"].count()
total_unique_schools
```

```python
#Total Students
total_students = students_df["Student ID"].count()
total_students
```

```python
#Total Budget
total_budget = schools_df["budget"].sum()
total_budget
```

```python
#Average Math Score
avg_math_score = round(students_df["math_score"].mean(),2)
avg_math_score
```

```python
#Average Reading Score
avg_reading_score = round(students_df["reading_score"].mean(),2)
avg_reading_score
```

```python
#Percentage Passing Math

passing_math = students_df["math_score"][students_df["math_score"] > 69].count()
percent_passing_math = format(round(passing_math/total_students,2),".0%")
percent_passing_math
```

```python
#Percentage Passing Reading

passing_reading = students_df["reading_score"][students_df["reading_score"] > 69].count()
percent_passing_reading = format(round(passing_reading/total_students,2),".0%")
percent_passing_reading
```

```python
#Percentage Passing
passing_score = 69
passing = (students_df["reading_score"][students_df["reading_score"] > passing_score]) * (students_df["math_score"][students_df["math_score"] > passing_score])

passing_total =format(round(passing.count()/total_students,2),".0%")

passing_total
```

```python
avg_math_scores = round(students_df.groupby(["school_name"])["math_score"].mean(),2)
avg_math_scores
```

```python
avg_reading_scores = round(students_df.groupby(["school_name"])["reading_score"].mean(),2)
avg_reading_scores
```

```python
#passing math (the percentage of students who passed math)

#condtion = (students_df["math_score"] > 69)
m_pass_count = students_df[students_df["math_score"] > 69].groupby("school_name")["math_score"].count()
m_total_count = students_df.groupby("school_name")["math_score"].count()

m_percent_pass = round(m_pass_count/total_count,2)

m_percent_pass
```

```python
# passing reading (the percentage of students who passed reading)

r_pass_count = students_df[students_df["reading_score"] > 69].groupby("school_name")["reading_score"].count()
r_total_count = students_df.groupby("school_name")["reading_score"].count()

r_percent_pass = round(r_pass_count/total_count,2)

r_percent_pass
```

```python
# overall passing (the percentage of students who passed math AND reading)

pass_count = students_df[(students_df["reading_score"] > 69) * (students_df["math_score"] > 69)].groupby("school_name")["reading_score"].count()
Overall_total_count = students_df.groupby("school_name")["Student ID"].count()

overall_pass = round(pass_count/Overall_total_count,2)

overall_pass
```
