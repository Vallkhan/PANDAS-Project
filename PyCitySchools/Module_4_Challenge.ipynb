```python
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")

passing_score = 69

#District unique schools
District_df = schools_df.groupby("type")["school_name"].count() 
District_df = District_df.reset_index()
District_df.rename(columns = {"school_name":"unique_schools"},inplace= True)

#District students
Dist_Stdnt = schools_df.groupby("type")["size"].sum() 
Dist_Stdnt = Dist_Stdnt.reset_index()
Dist_Stdnt = Dist_Stdnt.drop("type",axis=1)

#District budget
District_Budget = schools_df.groupby("type")["budget"].sum() 
District_Budget = District_Budget.reset_index()
District_Budget = District_Budget.drop("type", axis=1)

#District math average
District_math_score = students_df.groupby("school_name")["math_score"].mean() 
District_math_score = pd.merge(schools_df,District_math_score,on="school_name")
District_math_score = round(District_math_score.groupby("type")["math_score"].mean(),2)
District_math_score = District_math_score.reset_index()
District_math_score = District_math_score.drop("type", axis=1)

#District reading average
District_reading_score = students_df.groupby("school_name")["reading_score"].mean() 
District_reading_score = pd.merge(schools_df,District_reading_score,on="school_name")
District_reading_score = round(District_reading_score.groupby("type")["reading_score"].mean(),2)
District_reading_score = District_reading_score.reset_index()
District_reading_score = District_reading_score.drop("type",axis=1)

#Math percentage passing
total_count = students_df.groupby("school_name")["math_score"].count()
District_math_per = students_df[students_df["math_score"] > passing_score].groupby("school_name")["math_score"].count()/total_count 
District_math_per = pd.merge(schools_df,District_math_per,on="school_name")
District_math_per = round(District_math_per.groupby("type")["math_score"].mean(),2)
District_math_per = District_math_per.reset_index()
District_math_per = District_math_per.drop("type",axis=1)
District_math_per.rename(columns = {"math_score":"math_per"},inplace= True)
District_math_per.style.format({"math_per":"%{:.2f}"})

#Reading percentage passing
total_count = students_df.groupby("school_name")["reading_score"].count()
District_reading_per = students_df[students_df["reading_score"] > passing_score].groupby("school_name")["reading_score"].count()/total_count 
District_reading_per = pd.merge(schools_df,District_reading_per,on="school_name")
District_reading_per = round(District_reading_per.groupby("type")["reading_score"].mean(),2)
District_reading_per = District_reading_per.reset_index()
District_reading_per = District_reading_per.drop("type",axis=1)
District_reading_per.rename(columns = {"reading_score":"reading_per"},inplace= True)

#Over all percentage
total_count = students_df.groupby("school_name")["Student ID"].count() 
criteria = (students_df["reading_score"] > 69) * (students_df["math_score"]> 69)
District_passing_per = students_df[criteria].groupby("school_name")["Student ID"].count()/total_count
District_passing_per = pd.merge(schools_df,District_passing_per,on="school_name")
District_passing_per = round(District_passing_per.groupby("type")["Student ID"].mean(),2)
District_passing_per = District_passing_per.reset_index()
District_passing_per = District_passing_per.drop("type",axis=1)
District_passing_per.rename(columns = {"Student ID":"Math_Reading_Pass_Per"},inplace= True)




District_data_df = pd.concat([District_df,\
                            Dist_Stdnt,\
                            District_Budget,\
                            District_math_score,\
                            District_reading_score,\
                            District_math_per,\
                            District_reading_per,\
                            District_passing_per],axis = 1,join = "outer")

#Formatting
District_data_df['math_per'] = District_data_df['math_per'].map('{:.0%}'.format) 
District_data_df['reading_per'] = District_data_df['reading_per'].map('{:.0%}'.format)
District_data_df['Math_Reading_Pass_Per'] = District_data_df['Math_Reading_Pass_Per'].map('{:.0%}'.format) 
District_data_df['budget'] = District_data_df['budget'].map('${0:,.0f}'.format)
District_data_df
```

```python
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")

passing_score = 69

schools_df = schools_df.set_index("school_name")
schools_df = schools_df.sort_index(ascending=True)

#Budget per student
schools_df["budget_per_student"] = schools_df["budget"]/schools_df["size"] 
#budget_per_student = per_school_summary["budget_per_student"].astype("float").map("${:,.2f}".format)

#Average math score
avg_math_score_school = students_df.groupby("school_name")["math_score"].mean() 
avg_math_score_school = round(avg_math_score_school.reset_index(),2)
avg_math_score_school = avg_math_score_school.sort_index(ascending= True)


#Average reading score
avg_reading_score_school = students_df.groupby("school_name")["reading_score"].mean() 
avg_reading_score_school = round(avg_reading_score_school.reset_index(),2)
avg_reading_score_school = avg_reading_score_school.sort_index(ascending= True)

#Math percentage passing
total_count = students_df.groupby("school_name")["math_score"].count()
math_per = round(students_df[students_df["math_score"] > passing_score].groupby("school_name")["math_score"].count()/total_count,2) 
math_per = math_per.reset_index()
math_per = math_per.sort_index(ascending=True)

#Reading percentage passing
total_count = students_df.groupby("school_name")["reading_score"].count()
reading_per = round(students_df[students_df["reading_score"] > passing_score].groupby("school_name")["reading_score"].count()/total_count,2) 
reading_per = reading_per.reset_index()
reading_per = reading_per.sort_index(ascending=True)
#Overall passing
total_count = students_df.groupby("school_name")["Student ID"].count() 
criteria = (students_df["reading_score"] > 69) * (students_df["math_score"]> 69)
passing_per = round(students_df[criteria].groupby("school_name")["Student ID"].count()/total_count,2)
passing_per = passing_per.reset_index()
passing_per = passing_per.sort_index(ascending=True)
#passing_per

#schools_df = pd.concat([schools_df, avg_math_score_school],axis = 1,join = "outer")

per_school_summary = pd.merge(schools_df,avg_math_score_school,on="school_name")
per_school_summary = pd.merge(per_school_summary,avg_reading_score_school,on="school_name")
per_school_summary = pd.merge(per_school_summary,math_per,on="school_name")
per_school_summary = pd.merge(per_school_summary,reading_per,on="school_name")
per_school_summary = pd.merge(per_school_summary,passing_per,on="school_name")

per_school_summary.rename(columns = {"math_score_x":"avg_math_score"},inplace= True)
per_school_summary.rename(columns = {"reading_score_x":"avg_reading_score"},inplace= True)
per_school_summary.rename(columns = {"math_score_y":"math_per"},inplace= True)
per_school_summary.rename(columns = {"reading_score_y":"reading_per"},inplace= True)
per_school_summary.rename(columns = {"Student ID":"overall_passing_per"},inplace= True)

per_school_summary = per_school_summary.drop("School ID", axis= 1)
per_school_summary = per_school_summary.set_index("school_name")
per_school_summary["budget_per_student"] = per_school_summary["budget_per_student"].map('${0:,.0f}'.format)

per_school_summary['math_per'] = per_school_summary['math_per'].map('{:.0%}'.format)
per_school_summary['reading_per'] = per_school_summary['reading_per'].map('{:.0%}'.format)
per_school_summary['overall_passing_per'] = per_school_summary['overall_passing_per'].map('{:.0%}'.format)
# District_data_df['math_per'] = District_data_df['math_per'].map('{:.0%}'.format) 
# District_data_df['reading_per'] = District_data_df['reading_per'].map('{:.0%}'.format)
# District_data_df['Math_Reading_Pass_Per'] = District_data_df['Math_Reading_Pass_Per'].map('{:.0%}'.format) 
# District_data_df['budget'] = District_data_df['budget'].map('${0:,.0f}'.format)

per_school_summary
```

```python
per_school_summary_spending = per_school_summary
spending_math_scores = per_school_summary_spending.groupby("spending_range")["avg_math_score"].mean()
spending_math_scores = round(spending_math_scores.reset_index(),2)

spending_reading_scores = per_school_summary_spending.groupby("spending_range")["avg_reading_score"].mean()
spending_reading_scores = round(spending_reading_scores.reset_index(),2)
spending_reading_scores = spending_reading_scores.drop("spending_range",axis=1)

spending_passing_math = per_school_summary_spending.groupby("spending_range")["math_per"].mean()
spending_passing_math = round(spending_passing_math.reset_index(),2)
spending_passing_math = spending_passing_math.drop("spending_range",axis=1)

spending_passing_reading = per_school_summary_spending.groupby("spending_range")["reading_per"].mean()
spending_passing_reading = round(spending_passing_reading.reset_index(),2)
spending_passing_reading = spending_passing_reading.drop("spending_range",axis=1)

overall_passing_spending = per_school_summary_spending.groupby("spending_range")["overall_passing_per"].mean()
overall_passing_spending = round(overall_passing_spending.reset_index(),2)
overall_passing_spending = overall_passing_spending.drop("spending_range",axis=1)

Spending_Summary_df = pd.concat([spending_math_scores,\
                            spending_reading_scores,\
                            spending_passing_math,\
                            spending_passing_reading,\
                            overall_passing_spending],axis = 1,join = "outer")

Spending_Summary_df
```
